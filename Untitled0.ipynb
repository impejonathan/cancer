```python
import os
import random
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Load a model
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Select a random image from the images folder
images_folder = "/content/drive/MyDrive/PRO/Histogram10Grid4x-1/test/images"
image = random.choice(os.listdir(images_folder))
image_path = os.path.join(images_folder, image)

# Track with the model
results = model.track(source=image_path, show=True, tracker="bytetrack.yaml")

res = results[0].plot()
cv2_imshow(res)
```
